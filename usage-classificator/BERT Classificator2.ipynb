# BERT Classifier

## 1. Read the annotated data set

```python
import pandas as pd
```

```python
df = pd.read_csv('annotation_data_train_out.csv', sep=";")
df.drop(['used_Felix', 'Anmerkung'], axis=1, inplace=True)
df
```

```python
filtered_df = df[(df['used'] == "1") | (df['used'] == "0")]
print(len(filtered_df))
filtered_df
```

## 2. Train and Test data

```python
X = filtered_df[['ner', 'pre_sentence', 'sentence', 'post_sentence']],
y = pd.to_numeric(filtered_df['used'])
```

```python
X
```

## 3. Load SciBERT transformer

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModel, AutoConfig
```

```python
tokenizer = AutoTokenizer.from_pretrained('allenai/scibert_scivocab_uncased')
# Load pre-trained model (weights)
config = AutoConfig.from_pretrained("allenai/scibert_scivocab_uncased", output_hidden_states=True)
model = AutoModel.from_pretrained('allenai/scibert_scivocab_uncased', config=config)
```

```python
import torch
# from https://app.wandb.ai/cayush/bert-finetuning/reports/Sentence-classification-with-Huggingface-BERT-and-W%26B--Vmlldzo4MDMwNA
input_ids = []
attention_masks = []
sentences = [x['sentence'] for x in X]
for entry in filtered_df['sentence']:
    inputs = tokenizer.encode_plus(
            entry,
            add_special_tokens=True,
            return_tensors="pt",
            pad_to_max_length=True,
            max_length=512,
            return_attention_mask=True
        )
    input_ids.append(inputs['input_ids'])
    attention_masks.append(inputs['attention_mask'])

input_ids = torch.cat(input_ids, dim=0)
attention_masks = torch.cat(attention_masks, dim=0)
labels = torch.tensor([x for x in y])
```

```python
from torch.utils.data import TensorDataset, random_split

# Combine the training inputs into a TensorDataset.
dataset = TensorDataset(input_ids, attention_masks, labels)

# Calculate the number of samples to include in each set.
train_size = int(0.67 * len(dataset))
val_size = len(dataset) - train_size

# Divide the dataset by randomly selecting samples.
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
```

```python
from transformers import BertForSequenceClassification
model = BertForSequenceClassification.from_pretrained( 
        "allenai/scibert_scivocab_uncased", 
        num_labels = 2, 
        output_attentions = False, 
        output_hidden_states = False,
    )
```

```python
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler
batch_size = 8
train_dataloader = DataLoader(train_dataset,sampler = RandomSampler(train_dataset),  batch_size = batch_size)
validation_dataloader = DataLoader( val_dataset,  sampler = SequentialSampler(val_dataset), batch_size = batch_size)
```

```python
from torch.optim import AdamW
optimizer = AdamW(model.parameters(), lr = 1e-4, eps = 1e-8)
```

```python
device = torch.device('cuda')
model.to(device)
print(torch.cuda.memory_summary())
```

```python
import time
training_stats = []
total_t0 = time.time()
epochs = 20
for epoch_i in range(0, epochs):
    print('======== Epoch {:} / {:} ========'.format(epoch_i + 1, epochs))
    t0 = time.time()
    total_train_loss = 0
    model.train()
    for step, batch in enumerate(train_dataloader):
        if step % 40 == 0 and not step == 0:
            elapsed = time.time() - t0
            print('  Batch {:>5,}  of  {:>5,}.    Elapsed: {:}.'.format(step, len(train_dataloader), elapsed))
        b_input_ids = batch[0].to(device)
        b_input_mask = batch[1].to(device)
        b_labels = batch[2].to(device)
        model.zero_grad()        
        loss, logits = model(b_input_ids, 
                            token_type_ids=None, 
                            attention_mask=b_input_mask, 
                            labels=b_labels)

        total_train_loss += loss.item()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        del b_input_ids
        del b_input_mask
        del b_labels
    avg_train_loss = total_train_loss / len(train_dataloader)            
    training_time = time.time() - t0
    print("")
    print("  Average training loss: {0:.2f}".format(avg_train_loss))
```

```python
import numpy as np
total_eval_accuracy = []
for batch in validation_dataloader:
    b_input_ids = batch[0].cuda()
    b_input_mask = batch[1].to(device)
    b_labels = batch[2].to(device)
    with torch.no_grad():        
        (loss, logits) = model(b_input_ids, 
                              token_type_ids=None, 
                              attention_mask=b_input_mask,
                              labels=b_labels)


    logits = logits.detach()
    preds = torch.argmax(logits, dim=1).flatten()
    label_ids = b_labels
    m = (preds == label_ids).cpu().numpy().mean()
    total_eval_accuracy.append(m * 100)

avg_val_accuracy = np.mean(total_eval_accuracy)
print("  Accuracy: {0:.2f}".format(avg_val_accuracy))
```
